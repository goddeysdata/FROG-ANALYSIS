Importing the two datasets,both had equal numberof  rows and different number of columns with same primary key(gbifID ).To avoid repeating column names, specific column names for the second dataset(frog_data2) needed to be merged with the fist dataset(frog_data1).Due to this the two datasets  was possible to be joined using an inner join.

### MULTIVARIATE ANALYSIS

### Year against tmax and tmin

```python
final_frog.head(5)
```

```python
### checking all non null datasets to perform a multivariate ananlysis
final_frog=final_frog[(final_frog['eventDate'].notnull())&(final_frog['ppt_mean'].notnull())]
```

```python
final_frog.info()
```

```python
final_frog['year']=final_frog.eventDate.dt.year 
final_frog['year']=final_frog['year'].astype(int)
```

```python
final_frog.head(100)
```

### year_vs_temp

```python
year_vs_temp=final_frog.groupby('year')['tmax_mean','tmin_mean'].mean().reset_index()
year_vs_temp.head(5)
```

### year_vs_tmax

```python
plt.figure(figsize=[8,6])
plt.title("Year_vs_Max_Temperature",size =15)
plt.xlabel("Year",size =15)
plt.ylabel("Temperature",size = 16)
plt.yticks(size =15)
plt.plot(year_vs_temp['year'],year_vs_temp['tmax_mean'])
plt.show()
```

There was a downtrend around 1760 and a sharp uptrend around 1780.There was another sharp uptrend in the year1880,which is the maximum uptrend with temperature value of 34.0.The lowest temperature was around 18.9. This means in the year 1760 most of the frogs were seen around hot temperatures with the value 26.9,around 1780 most  of the frogs were seen at a more slightly warm environment.In the year 1880 most of the frogs were seen in very hot environment with a temperature value of 34.0.Looking at the difference between the lowest and maximum temperature this meens most of the frogs were seen in hot environment.

### year_vs_tmin

```python
plt.figure(figsize=[8,6])
plt.title("Year_vs_Min_Temperature",size =15)
plt.xlabel("Year",size =15)
plt.ylabel("Temperature",size = 16)
plt.yticks(size =15)
plt.plot(year_vs_temp['year'],year_vs_temp['tmin_mean'])
plt.show()
```

There was a downtrend around 1760 and a sharp uptrend around 1810.There was another sharp uptrend in the year 1920,which is the maximum uptrend with temperature value of 22.0.The lowest temperature value is around 6.9. This means in the year 1760 most of the frogs were seen around slightly cold temperatures with a value of 15.9,around 1810 most of the frogs were seen at cold environment.In the year 1930 most of the frogs were seen in a slightly warm environment with a temperature value of 22.0.Looking at the difference between the lowest and maximum temperature this meens most of the frogs were seen in colder environment.
